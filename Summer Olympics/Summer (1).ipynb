# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
cities = df["City"].unique() #unique cities in an array
no_of_cities = len(cities)
print("In {} cities summmer olympics is held so far".format(no_of_cities))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
data = []
for sport in df["Sport"].unique() :
    df = df[df["Sport"] == sport]
    data.append([sport,len(df[df["Medal"] == "Gold"])]) #appending sport name and no of gold medals in that sport
    df = pd.read_csv("summer.csv")
data = pd.DataFrame(data , columns = ["Sport" , "No of Gold Medals"]) #framing data
data = data.sort_values(by = "No of Gold Medals",ascending = False).head() # sorting with respect to no of medals for top five 
data = data.plot(x = "Sport",y = "No of Gold Medals",kind = "bar",figsize = (10,5)) # plotting bar graph 
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
data = []
for sport in df["Sport"].unique() :
    df = df[df["Sport"] == sport]
    data.append([sport, len(df[df["Sport"] == sport])]) # appending sport and total no of medals is equal to no of rows in the sport 
    df = pd.read_csv("summer.csv")
data = pd.DataFrame(data , columns = ["Sport" , "No of Medals"]) # framing data 
data = data.sort_values(by = "No of Medals",ascending = False).head()
data = data.plot(x = "Sport",y = "No of Medals",kind = "bar",figsize = (10,5)) #plotting graph for top five sport
```

### 4. Which player has won most number of medals? (Top 5)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
data = []
players_array = df["Athlete"].unique()
for player in players_array :
    data.append([player, len(df[df["Athlete"] == player])])
data = pd.DataFrame(data , columns = ["Player" , "No of Medals"]).sort_values(by = "No of Medals" , ascending = False).head()
data.plot(x = "Player" , y = "No of Medals" , kind = "bar" , figsize = (10,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
data = []
players_array = df["Athlete"].unique() #players list
for player in players_array :
    player_df = df[df["Athlete"] == player] # unique player dataframe
    data.append([player, len(player_df[player_df["Medal"] == "Gold"])]) #appending data with data and length of golg medals
data = pd.DataFrame(data , columns = ["Player" , "No of Gold Medals"]).sort_values(by = "No of Gold Medals" , ascending = False).head()
data.plot(x = "Player" , y = "No of Gold Medals" , kind = "bar" , figsize = (10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
df = df[df["Country"] == "IND"].sort_values(by = "Year")
df = df[df["Medal"] == "Gold"].head(1)
year = int(df["Year"])
print("In {} India won first gold medal in summer olympics".format(year))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
data = []
for event in df["Event"].unique() :
    data.append([event , len(df[df["Event"] == event])])
data = pd.DataFrame(data , columns = ["Event" , "No of Players"]).sort_values(by = "No of Players" , ascending = False).head()
data = data.plot(x = "Event" , y = "No of Players" , kind = "bar" , figsize = (10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
import pandas as pd
import numpy as np
df = pd.read_csv("summer.csv")
data = []
for sport in df["Sport"].unique() :
    player_df = df[df["Sport"] == sport]
    female_df = player_df[player_df["Gender"] == "Women"]
    gold_medal_df = female_df[female_df["Medal"] == "Gold"]
    data.append([sport,len(gold_medal_df)])
data = pd.DataFrame(data , columns = ["Sport" , "no of women"]).sort_values(by = "no of women" , ascending = False).head()
data.plot(x = "Sport",y = "no of women",kind = "bar",figsize = (10,5))
```
